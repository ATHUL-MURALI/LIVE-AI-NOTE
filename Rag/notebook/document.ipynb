```python
sample_text = {
    "../data/text_files/python_intro.txt" : """Python is a high-level, interpreted programming language known for its simplicity, readability, and versatility. Created by Guido van Rossum and first released in 1991, Python emphasizes clear syntax and code readability, making it an excellent choice for beginners while remaining powerful for professionals. It supports multiple programming paradigms, including procedural, object-oriented, and functional programming. Python’s extensive standard library and vast ecosystem of third-party packages enable it to be used in diverse fields such as web development, data science, artificial intelligence, automation, and scientific computing. Its cross-platform nature and active community have made Python one of the most popular and widely used programming languages in the world.""" ,
    "../data/text_files/machine_learning.txt" : """Machine Learning (ML) is a subset of Artificial Intelligence (AI) that focuses on building algorithms capable of learning patterns from data and making predictions or decisions without being explicitly programmed for each task. Instead of following hard-coded rules, ML models improve their performance through experience—adapting as they process more data.

At its core, ML involves training a model on historical data so it can generalize to unseen data. This generalization is the ultimate goal, enabling applications like fraud detection, recommendation systems, speech recognition, and autonomous driving.

Main Types of Machine Learning

Supervised Learning – Uses labeled datasets to map inputs to known outputs. Common for classification (e.g., spam detection) and regression (e.g., price prediction).

Unsupervised Learning – Works with unlabeled data to find hidden patterns, clusters, or associations (e.g., customer segmentation).

Semi-Supervised Learning – Combines a small amount of labeled data with large amounts of unlabeled data to improve learning efficiency.

Reinforcement Learning – Trains an agent to make sequential decisions by rewarding desirable actions and penalizing undesirable ones (e.g., robotics, game AI).

How ML Works

Data Preparation: Collect, clean, and preprocess data (feature engineering, normalization).

Model Selection: Choose an algorithm (e.g., decision trees, neural networks).

Training: Adjust model parameters to minimize error (loss function optimization).

Evaluation: Test on unseen data to measure accuracy and avoid overfitting.

Deployment: Integrate into real-world systems for inference.

Example: Simple Linear Regression in Python

from sklearn.linear_model import LinearRegression
import numpy as np

# Training data: square footage vs. price
X = np.array([[1500], [2000], [2500], [3000]]) # Features
y = np.array([300000, 400000, 500000, 600000]) # Labels

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict price for 2800 sq ft
predicted_price = model.predict([[2800]])
print(f"Predicted Price: ${predicted_price[0]:,.2f}")
Copy
This example demonstrates supervised learning where the model learns a mapping from house size to price.

Key Applications

Computer Vision: Object detection, medical imaging.

Natural Language Processing: Chatbots, translation, sentiment analysis.

Predictive Analytics: Stock forecasting, demand prediction.

Generative AI: Text, image, and audio generation.

Considerations While ML offers automation, personalization, and efficiency, it also poses challenges like bias in training data, lack of interpretability, and ethical concerns. Proper MLOps practices ensure models remain accurate, fair, and reliable over time.

Would you like me to create a visual diagram showing how supervised, unsupervised, and reinforcement learning differ? That could make the distinctions even clearer.""" 
}

for file_path , content in sample_text.items():
    with open(file_path, 'w', encoding="utf-8") as f:
        f.write(content)
print("sample txt content created")
```
